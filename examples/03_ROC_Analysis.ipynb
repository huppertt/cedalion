```python
import cedalion.testing.channelROC as roc

import cedalion.testing.simData as simdata
import cedalion.pipelines.modules.preproccessing as prep
import cedalion.pipelines.modules.glm as stats
```

```python
model=roc.ChannelROC()
model.data_simulation_function=simdata.Data
model.pipeline = roc.PipelineList([prep.intensity_opticaldensity,
                                    prep.mbll,
                                    prep.resample,
                                    stats.GLM])
model.data_simulation_args={'snr':5}  # These are passed to the simulation function
model.pipeline_args={'noise_model':'ar_irls','Fs':4,'ar_order':32} # These are set on the pipeline object
```

```python
model.run(10)
# run incremental number of simulations
```

```python
model.draw()
```
